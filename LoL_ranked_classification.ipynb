```python
import pandas as pd
```

```python
df = pd.read_csv('high_diamond_ranked_10min.csv')
```

```python
df.head()
```

```python
X = df.drop(['gameId', 'blueWins'], axis=1).values
y = df['blueWins'].values
```

```python
df = df.sample(frac=1).reset_index(drop=True)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
import torch 
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
```

```python
X_train = torch.FloatTensor(X_train)
X_test = torch.FloatTensor(X_test)
y_train = torch.LongTensor(y_train)
y_test = torch.LongTensor(y_test)
```

```python
df['blueWins'].value_counts()
```

```python
class TabularModel(nn.Module):
    def __init__(self, n_cont, out_sz, layers, p=0.5):
        super().__init__()
        #self.bn_cont = nn.BatchNorm1d(n_cont)
        
        layerlist = []
        n_in = n_cont
        
        for i in layers:
            layerlist.append(nn.Linear(n_in, i))
            layerlist.append(nn.ReLU(inplace=True))
            #layerlist.append(nn.BatchNorm1d(i))
            layerlist.append(nn.Dropout(p))
            n_in = i
            
        layerlist.append(nn.Linear(layers[-1], out_sz))
        
        self.layers = nn.Sequential(*layerlist)
        
    def forward(self, x):
        #x = self.bn_cont(x)
        x = self.layers(x)
        
        return x
```

```python
model = TabularModel(len(X[1]), 2, [200, 100], p=0.4)
```

```python
model
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 1000
losses = []
```

```python
for i in range(epochs):
    y_pred = model.forward(X_train)
    loss = criterion(y_pred, y_train)
    losses.append(loss)
    
    if i % 50 == 0:
        print(f'epoch: {i} loss: {loss}')
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
plt.plot(range(epochs), losses)
```

```python
total = 0
correct = 0

with torch.no_grad():
    for i, data in enumerate(X_test):
        y_val = model.forward(data)
        
        predicted = y_val.argmax().item()
        true = y_test[i].item()
        print(f'PREDICTED CLASS: {predicted} TRUE CLASS: {true}')
        
        if predicted == true:
            correct += 1
        
        total += 1
        
        
print(correct)
print(total)
print(correct/total)
        
```
